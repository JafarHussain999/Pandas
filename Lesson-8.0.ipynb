# Grouping

```python
country_group = df.groupby(["Country"])
```

```python
country_group.get_group("India")
```

```python
country_group["OpSys"].value_counts().loc["India"] #let see how many users use the OS in india
```

```python
country_group["Salary"].median()
```

```python
country_group["Salary"].median().loc["India"]
```

# Aggregation 

```python
country_group["Salary"].agg(['median','mean'])
```

```python
country_group["Salary"].agg(['median','mean']).loc["India"]
```

```python
filt = df["Country"] == 'India'
df.loc[filt]['LanguageWantToWorkWith'].str.contains('Python').sum()
```

```python
country_group['LanguageWantToWorkWith'].apply(lambda x : x.str.contains('Python').sum()) #how many know python
```

```python
Total_No_of_people_from_country = df['Country'].value_counts()
Total_No_of_people_from_country
```

```python
Total_No_of_people_Whou_knows_Python = country_group['LanguageWantToWorkWith'].apply(lambda x : x.str.contains('Python').sum())
Total_No_of_people_Whou_knows_Python
```

```python
python_df = pd.concat([Total_No_of_people_from_country,Total_No_of_people_Whou_knows_Python],axis='columns')
python_df
```

```python
python_df.rename(columns = {"Country" : "Total_No_of_people_from_country" , "LanguageWantToWorkWith" : "Total_No_of_people_Whou_knows_Python"},inplace=True)
python_df 
```

```python
python_df["Percentage_Know_Python"] = (Total_No_of_people_Whou_knows_Python / Total_No_of_people_from_country ) * 100 #here we calculate the percentage of the people know python language by the country
```

```python
python_df
```

```python
python_df.sort_values(by='Percentage_Know_Python',inplace=True)
python_df
```
